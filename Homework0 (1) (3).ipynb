```python
# STEP 1: Download the tweets (posts) of “EarlyBirdLearn” 
userID = "EarlyBirdLearn"
tweets=[]
for tweet in tweepy.Cursor(api.user_timeline, id=userID, include_rts=True, tweet_mode='extented').items(5000):
    tweets.append(tweet)

#Write the json objects in a file named EarlyBirdLearn-tweets.json
with open('EarlyBirdLearn-tweets.json', 'w') as fp:
    for tweet in tweets:
        json.dump(tweet._json, fp)
        fp.write('\n')
        
#Extract the text of tweets and save them in file named EarlyBirdLearn-tweets-text.txt        
with open('EarlyBirdLearn-tweets-text.txt', 'w') as fp:
    for tweet in tweets:
        json.dump(tweet.text, fp)
        fp.write('\n')
```
